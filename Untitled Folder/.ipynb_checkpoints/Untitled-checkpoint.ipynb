```python
import time
from threading import Thread
import cv2
import numpy as np
import serial
import sys
from curtsies import Input

_USE_VIDEO_FILE = 0

video_file = 'track-s.mkv'
if _USE_VIDEO_FILE :
	cap = cv2.VideoCapture(video_file)
else : 
	cap = cv2.VideoCapture(-1)
	
ser = serial.Serial(port='/dev/ttyUSB0',baudrate=9600)
Angle = 0
MotorPower = 0  

#Motor Control - angle
class WheelHandler():
    global ser

    #ser=0
#    def __init__(self):
#        self.ser=serial.Serial(port='/dev/ttyUSB0',baudrate=9600)

    def Forward(self):
        ser.write(bytes("!F".encode('ascii')))

    def Backward(self):
        ser.write(bytes("!B".encode('ascii')))

    def Stop(self):
        ser.write(bytes("!S".encode('ascii')))

    def XL(self):
        ser.write(bytes("!XL".encode('ascii')))

    def XR(self):
        ser.write(bytes("!XR".encode('ascii')))

    def ZC(self):
        ser.write(bytes("!ZC".encode('ascii')))
        
    def ZD(self):
        ser.write(bytes("!ZD".encode('ascii')))

    def ZE(self):
        ser.write(bytes("!ZE".encode('ascii')))

    def ZF(self):
        self.ser.write(bytes("!ZF".encode('ascii')))

    def AA(self):
        ser.write(bytes("!AA".encode('ascii')))

    def TEST(self):
        self.ser.write(bytes("040040040040#".encode('ascii')))

    def TEST1(self):
        self.ser.write(bytes("#".encode('ascii')))

    def Wheel_Clean(self):
        self.ser.close()


def Input_key(k,wheel):
    '''     keyboard input function     '''
    print(k)
    if k==u']':
        wheel.ser.close()
        sys.exit(1)
    #elif k==u'a':
        #wheel.XL()
    else:
        print("Press Wheel Control Key or  Exit KEY(])")
#Image Processing Lane Recognition

def Grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def Gaussian_blur(img):
    return cv2.GaussianBlur(img, (5, 5), 0)

def Canny(img, low_threshold, high_threshold):
    return cv2.Canny(img, low_threshold, high_threshold)

def Region_of_interest(img, vertices, vertices2):
    mask = np.zeros_like(img)
    if len(img.shape) > 2:
        channel_count = img.shape[2]
        mask_color = (255,) * channel_count
    else:
        mask_color = 255
        cv2.fillPoly(mask, vertices, mask_color)
        cv2.fillPoly(mask, vertices2, mask_color)
        masked_img = cv2.bitwise_and(img, mask)
    return masked_img

def HoughTransform(img, threshold):
    lines = cv2.HoughLines(img, 1, np.pi / 180, threshold)
    #print(lines.shape)
    l = []
    if lines is not None:
        for i in range(len(lines)):
            for rho, theta in lines[i]:
                a = np.cos(theta)
                b = np.sin(theta)
                x0 = a * rho
                y0 = b * rho
                x1 = int(x0 + 1000 * (-b))
                y1 = int(y0 + 1000 * (a))
                x2 = int(x0 - 1000 * (-b))
                y2 = int(y0 - 1000 * (a))

                l.append([[x1, y1, x2, y2]])
    l = np.squeeze(np.array(l))
    return l

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):  # hough transform
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len,
                            maxLineGap=max_line_gap)
    l = []

    for line in lines:
        l.append(line[0])
        l = np.squeeze(np.array(l))
    return l


def Draw_line(img, line, color=[0, 0, 255], thickness=2):
    x1, y1, x2, y2 = line
    cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def GetIntersection(line1, line2):
    line1 = [[line1[0], line1[1]], [line1[2], line1[3]]]
    line2 = [[line2[0], line2[1]], [line2[2], line2[3]]]
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
        return
    d = (det(*line1), det(*line2))
    x = det(d, xdiff) // div
    y = det(d, ydiff) // div
    return (x, y)

def mouse_callback(event, x, y, flags, param): 
    print("mouse event, x:", x ," y:", y) 

# set ROI Boxes
ROI_Top= 250
ROI_Height = 120
ROI_LEFT = 150
ROI_RIGHT = 230

def Double_line(dist):
    degree = dist 
    if degree > 50: degree = 50
    if degree < -50: degree = -50
    return degree

def Single_line(dist, handle):
    global ROI_RIGHT
    global ROI_LEFT
    angle = 0
    if handle:
        print("RIGHT")
        if dist > 20:
            #angle = -50
            angle = -45 - (dist*20)//ROI_RIGHT
        else:
            angle = -30
    else:
        print("LEFT")
        if dist > 20:
            #angle = 50
            angle = 45 + (dist*20)//ROI_LEFT
        else:
            angle = 30
    return angle


def thread_run():
    global Angle
    global exitFlg
    time.sleep(0.5)
    packet = b''
    while True:
        #print("thread2 on")
        FORWARD_SPEED = 40
        TURN_SPEED = 60
        TURN_DIFF = 2
        SPEED_LIMIT = 60
		# Set Wheel Speed 
		# REAR
		# 0 1
		# 2 3
		# FRONT 
		# [0]#[1]$[2]%[3]^
        if abs(Angle)< 40 :          
            FORWARD_SPEED = FORWARD_SPEED+500
            packet = bytes("{0:03d}#{1:03d}${2:03d}%{3:03d}^".format(FORWARD_SPEED,FORWARD_SPEED,FORWARD_SPEED,FORWARD_SPEED).encode('ascii'))
            ser.write(packet)
            time.sleep(0.01)
            #wheel.Forward()
            wheel.AA()
        elif Angle > 45 : # Left
            #ser.write(bytes("045#055$045%055^".encode('ascii')))
            TURN_SPEED = TURN_SPEED + (Angle-45)
            if TURN_SPEED > SPEED_LIMIT :
                TURN_SPEED =SPEED_LIMIT 
            elif TURN_SPEED < 0: 
            	TURN_SPEED =0
            packet = bytes("{0:03d}#{1:03d}${2:03d}%{3:03d}^".format(500-TURN_SPEED+TURN_DIFF,TURN_SPEED+500,500-TURN_SPEED+TURN_DIFF,500+TURN_SPEED).encode('ascii'))
            print("LEFT : {0}".format(TURN_SPEED))
            ser.write()
            time.sleep(0.01)
            #wheel.ZD() 
            wheel.AA()
        elif Angle < -45: # Right  
            #ser.write(bytes("055#045$055%045^".encode('ascii')))
            TURN_SPEED = TURN_SPEED - (45+Angle)
            if TURN_SPEED > SPEED_LIMIT  :
                TURN_SPEED = SPEED_LIMIT
            elif TURN_SPEED < 0: 
            	TURN_SPEED =0
            print("RIGHT : {0}".format(TURN_SPEED))
            packet = bytes("{0:03d}#{1:03d}${2:03d}%{3:03d}^".format(TURN_SPEED+500,500-TURN_SPEED+TURN_DIFF,500+TURN_SPEED,500-TURN_SPEED+TURN_DIFF).encode('ascii'))
            ser.write(packet)
            time.sleep(0.01)
            #wheel.ZC()
            wheel.AA()
        time.sleep(0.03)
        print(packet)
        if exitFlg:
            ser.write(bytes("000#000$000%000^".encode('ascii')))
            time.sleep(0.01)
            wheel.Stop()
            time.sleep(0.01)
            break;
		

def Start():
    global Angle
    global MotorPower
    global cap
    global exitFlg
    #wheel.TEST1()
    #wheel.Forward()
 
    #print("Thread2 on")
    cv2.namedWindow("img")   # create a named window
    cv2.moveWindow("img", 340, 30)   # Move it to (40, 30)
    while True:
        ret, img = cap.read()
        if ret == False :
            print("Failed");
            continue
        shape = img.shape
        vertices = np.array([[(0, shape[0]-ROI_Height), (0, ROI_Top), (ROI_LEFT,ROI_Top), (ROI_LEFT, shape[0]-ROI_Height)]], dtype=np.int32)
        vertices2 = np.array([[(shape[1]-ROI_RIGHT, shape[0]-ROI_Height), (shape[1]-ROI_RIGHT, ROI_Top), (shape[1], ROI_Top), (shape[1], shape[0]-ROI_Height)]], dtype=np.int32)

        try:
            pointX_left, pointX_right = [None, None]
            gray = Grayscale(img)
            blur = Gaussian_blur(gray)
            edges = Canny(blur, 50, 150)


            #vertices = np.array([[(0, shape[0]), (0, 350), (120, 350), (230, shape[0])]], dtype=np.int32)
            #vertices2 = np.array([[(530, shape[0]), (530, 350), (shape[1], 350), (shape[1], shape[0])]], dtype=np.int32)
            masked_img = Region_of_interest(edges, vertices, vertices2)
            line_arr = HoughTransform(masked_img, 50)
            slope_degree = (np.arctan2(line_arr[:, 1] - line_arr[:, 3], line_arr[:, 0] - line_arr[:, 2]) * 180) / np.pi
			#SLOPE_LIMIT_HIGH = 160
			#SLOPE_LIMIT_LOW = 95

            SLOP_LIMIT_HIGH  = 160
            SLOP_LIMIT_LOW = 95

            #line_arr = line_arr[np.abs(slope_degree) < SLOPE_LIMIT_HIGH]
            #slope_degree = slope_degree[np.abs(slope_degree) < SLOPE_LIMIT_HIGH]
            #line_arr = line_arr[np.abs(slope_degree) > SLOPE_LIMIT_LOW]
            #slope_degree = slope_degree[np.abs(slope_degree) > SLOPE_LIMIT_LOW]

            line_arr = line_arr[np.abs(slope_degree) < 180]
            slope_degree = slope_degree[np.abs(slope_degree) < 180]
            line_arr = line_arr[np.abs(slope_degree) > 90]
            slope_degree = slope_degree[np.abs(slope_degree) > 90]
            L_lines, R_lines = line_arr[(slope_degree > 0), :], line_arr[(slope_degree < 0), :]
            line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
            standard_line = [0, ROI_Top, shape[1], ROI_Top]
            standard_line2 = [0,shape[0]-ROI_Height,shape[1],shape[0]-ROI_Height]
            Draw_line(line_img, standard_line)
            Draw_line(line_img, standard_line2)

            if L_lines.shape[0] != 0:
                line = np.mean(L_lines, axis=0, dtype=np.int32)
                Draw_line(line_img, line)
                intersection_point = GetIntersection(line, standard_line)
                if intersection_point:
                    pointX_left = intersection_point[0]
                    cv2.circle(line_img, intersection_point, 5, (0, 0, 255), 3)
            if R_lines.shape[0] != 0:
                line = np.mean(R_lines, axis=0, dtype=np.int32)
                Draw_line(line_img, line)
                intersection_point = GetIntersection(line, standard_line)
                if intersection_point:
                    pointX_right = intersection_point[0]
                    cv2.circle(line_img, intersection_point, 5, (0, 0, 255), 3)

            if pointX_right and pointX_left:
                dist = (pointX_right + pointX_left) // 2 - shape[1] // 2
                Angle = Double_line(dist)
            else:
                if pointX_right:
                    Angle = Single_line(shape[1] - pointX_right, True)
                if pointX_left:
                    Angle = Single_line(pointX_left, False)

			# Draw ROI Boxes
            angleStr = "Angle : {0}".format(Angle)
            try:
                cv2.putText(img,angleStr,(200,ROI_Top-50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0))
                pass
            except:
                print("Text Err")
            cv2.polylines(line_img,vertices,True,(0,255,0),2)
            cv2.polylines(line_img,vertices2,True,(0,255,0),2)
			#cv2.polylines(line_img,vertices3,True,(255,0,0),2)
			#cv2.polylines(masked_img2,vertices3,True,(255,0,0),2)

            print(angleStr)

            cv2.setMouseCallback('img', mouse_callback)         
            img = cv2.addWeighted(img, 1, line_img, 1, 0)
            #cv2.imshow('blur', blur)
            cv2.imshow("img", img)
            if cv2.waitKey(10) & 0xFF == ord('q'):
                exitFlg = True;
                wheel.Stop()
                break
        except:
            print("err")
            line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
            cv2.polylines(line_img,vertices,True,(0,255,0),2)
            cv2.polylines(line_img,vertices2,True,(0,255,0),2)
            img = cv2.addWeighted(img, 1, line_img, 1, 0)
            cv2.imshow("img", img)
            if cv2.waitKey(10) & 0xFF == ord('q'):
                exitFlg = True;
                wheel.Stop()
                break
            continue

    cap.release()
    cv2.destroyAllWindows()

if __name__ == '__main__':
    wheel=WheelHandler()
    exitFlg = False;
    t1 = Thread(target = Start)
    t2 = Thread(target = thread_run)
    t1.start()
    t2.start()
    if exitFlg:
        t1.join()
        t2.join()
```
